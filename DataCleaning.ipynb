##### Identity

- X1SEX - Student's sex 
- X1HISPANIC, X1WHITE, X1BLACK, X1ASIAN, X1PACISLE, X1AMINDIAN - Student's race
- X1LOCALE - School's urbanicity
- X1REGION - School's Region

```python
id_data = data[['X1SEX', 'X1HISPANIC', 'X1WHITE', 'X1BLACK',
                'X1ASIAN', 'X1PACISLE', 'X1AMINDIAN', 
                'X1LOCALE', 'X1REGION']]
count(id_data)
```

Because this is sensitive, potentially identifying information some of the race data was suppressed from the publicly available data. -9 is a placeholder for missing values, and -5 indicates that the data was suppressed.

```python
data = data[data['X1SEX']!=-9]
for col in ['X1HISPANIC', 'X1WHITE', 'X1BLACK']:
    data.loc[data[col]==-9, col]= 0
```

```python
data.drop(['X1ASIAN', 'X1PACISLE', 'X1AMINDIAN'], axis = 1,
          inplace = True)
```

```python
data.loc[data['X1LOCALE']== 1, 'X1LOCALE'] = 'City'
data.loc[data['X1LOCALE']== 2, 'X1LOCALE'] = 'Suburb'
data.loc[data['X1LOCALE']== 3, 'X1LOCALE'] = 'Town'
data.loc[data['X1LOCALE']== 4, 'X1LOCALE'] = 'Rural'
```

```python
data.loc[data['X1REGION']== 1, 'X1REGION'] = 'Northeast'
data.loc[data['X1REGION']== 2, 'X1REGION'] = 'Midwest'
data.loc[data['X1REGION']== 3, 'X1REGION'] = 'South'
data.loc[data['X1REGION']== 4, 'X1REGION'] = 'West'
```

##### Background
- X1PAREDU - Highest Parent Education Levels
- X1PARPATTERN - Parent Relationship
- X1HHNUMBER - Number of Household Members 
- X1FAMINCOME, X1POVERTY, X1POVERTY130, X1POVERTY185 - Income Status

```python
bg_data = data[['X1PAREDU', 'X1PARPATTERN', 
                'X1HHNUMBER', 'X1FAMINCOME', 'X1POVERTY', 
                'X1POVERTY130', 'X1POVERTY185', 
                'X4PARDATE']]
count(bg_data)
```

According to the codebook, -9 indicates missing values and -8 indicates non-response. Due to the nature of the information we are gathering, in both cases the rows containing these values will be removed.

```python
for col in bg_data.columns:
    data.loc[data[col]==-7, col]= 1000
    data = data[data[col] >= 0]
    data.loc[data[col]==1000, col]= 'N/A'
```

```python
for col in ['X1PAREDU']:
    data.loc[data[col]== 1, col]= 'Less than high school'
    data.loc[data[col]== 2, col]= 'HS diploma or GED'
    data.loc[data[col]== 3, col]= 'Associates degree'
    data.loc[data[col]== 4, col]= 'Bachelors degree'
    data.loc[data[col]== 5, col]= 'Masters degree'
    data.loc[data[col]== 7, col]= 'high lvl prof degree'
```

```python
for col in ['X1PARPATTERN']:
    data.loc[data[col]== 1, col]= 'Two bio/adoptive parents'
    data.loc[data[col]== 2, col]= 'Bio/adoptive mom + other'
    data.loc[data[col]== 4, col]= 'Bio/adoptive dad + other'
    data.loc[data[col]== 6, col]= 'Two other guardians'
    data.loc[data[col]== 7, col]= 'Bio/adoptive mom only'
    data.loc[data[col]== 8, col]= 'Bio/adoptive dad only'
    data.loc[data[col]== 9, col]= 'female guardian only'
    data.loc[data[col]== 10, col]= 'male guardian only'
    data.loc[data[col]== 11, col]= 'P1/P2 less than half'
```

```python
for col in ['X1FAMINCOME']:
    data.loc[data[col]== 1, col]= '< $15k'
    data.loc[data[col]== 2, col]= '> $15k and <= $35k'
    data.loc[data[col]== 3, col]= '> $35k and <= $55k'
    data.loc[data[col]== 4, col]= '> $55k and <= $75k'
    data.loc[data[col]== 5, col]= '> $75k and <= $95k'
    data.loc[data[col]== 6, col]= '> $95k and <= $115k'
    data.loc[data[col]== 7, col]= '> $115k and <= $135k'
    data.loc[data[col]== 8, col]= '> $135k and <= $155k'
    data.loc[data[col]== 9, col]= '> $155k and <= $175k'
    data.loc[data[col]== 10, col]= '> $175k and <= $195k'
    data.loc[data[col]== 11, col]= '> $195k and <= $215k'
    data.loc[data[col]== 12, col]= '> $215k and <= $235k'
    data.loc[data[col]== 13, col]= '> $235k'
```

```python
data['X4PARDATE'].value_counts()
```

```python
for col in ['X4PARDATE']:
    data.loc[data[col] == 'N/A', 'Parent?']= 'Not a Parent'
    data.loc[data[col] == 201300, 'Parent?']= 'During HS'
    data.loc[data[col] == 201200, 'Parent?']= 'During HS'
    data.loc[data[col] == 201100, 'Parent?']= 'During HS'
    data.loc[data[col] == 201600, 'Parent?']= 'After HS'
    data.loc[data[col] == 201500, 'Parent?']= 'After HS'
    data.loc[data[col] == 201400, 'Parent?']= 'After HS'
data.drop('X4PARDATE', axis=1, inplace=True)
```

##### College Services

```python
coll_data = data[['S3CNSLCLG', 'S3CNSLAID', 
                  'S1TALENTSRCH', 'S1UPWARDBND', 
                  'S1GEARUP', 'S1AVID', 'S1MESA', 
                  'C1ENCCLG', 'C1CLGFAIR', 
                  'C1VISITCLG', 'C1INFOSESSN', 
                  'C1FINANCEAID']]
count(coll_data)
```
